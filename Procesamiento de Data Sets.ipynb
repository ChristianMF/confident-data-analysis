```python
#1-M-19NOV-0800-0835-3A
df1M = spark.read.options(inferSchema=True, multiLine=True, delimiter="},", header=False).json("/Users/christian/Desktop/Data/1-M-19NOV-0800-0835-3A.json")
```

```python
#Normalizaci√≥n 1-M-19NOV-0800-0835-3A
df1M=spark.createDataFrame(df1M.collect()[0])
a=[]
b=[]
c=[]
d=[]
e='3A'
def normaliza(data_structure):
    f=1
    for (index, item) in enumerate(data_structure):
        a.append(data_structure.collect()[0][index][0])
        b.append(data_structure.collect()[0][index][1])
        c.append(f)
        d.append(e)
        
        f+=1
    return data_structure

g=[]
def normalizaWithDate(data_structure):
    h=0
    for row in data_structure.collect():
        g.append(data_structure.select(data_structure.user.substr(23,43)).collect()[h][0])
        h+=1
    return data_structure

dftest=normaliza(df1M)
df1M=spark.createDataFrame(zip(c, a, b, d), schema=['no','text','user','group'])

dftest=normalizaWithDate(df1M)
df1M=spark.createDataFrame(zip(c, a, b, g, d), schema=['no','text','user','date','group'])

df1M.show()
df1M.summary().show()
```
